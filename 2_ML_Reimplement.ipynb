```python
import pandas as pd
```

```python
ddos_file_path = '/content/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv'

home_data = pd.read_csv(ddos_file_path)
```

```python
  home_data.describe()
```

```python
home_data.head()
```

```python
home_data.columns
```

```python
home_data = home_data.replace(to_replace = "BENIGN", value = 0)
```

```python
home_data = home_data.replace(to_replace = "DDoS", value = 1)
```

```python
home_data.head()
```

```python
home_data = home_data.replace(to_replace = "Infinity", value = 0)
```

```python
home_data.describe()
```

```python
home_data['Flow Bytes/s'].isnull().values.any()
```

```python
home_data = home_data.fillna(home_data.median())
```

```python
home_data['Flow Bytes/s'].isnull().values.any()
```

```python
y= home_data[' Label']
```

```python
from sklearn.feature_selection import SelectPercentile, chi2, f_classif
selector = SelectPercentile(f_classif, percentile=15)
# selector.fit(X, y)
```

```python
X = home_data.drop(' Label', axis = 1)
```

```python
X.head()
```

```python
X = X.drop('Flow ID', axis = 1)
```

```python
X.head()
```

```python
X = X.drop(' Source IP', axis = 1)
```

```python
X = X.drop(' Destination IP', axis = 1)
```

```python
X.head()
```

```python
X = X.drop(' Timestamp', axis = 1)
```

```python
X.head()
```

```python
X = X.reset_index()
```

```python
X.head()
```

```python
X.fillna(X.mean())
```

```python
import numpy as np
np.any(np.isnan(X))
```

```python
np.all(np.isfinite(X))
```

```python
np.any(np.isnan(y))
```

```python
np.all(np.isfinite(y))
```

```python
X.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
X.fillna(X.median(), inplace = True)
```

```python
np.any(np.isnan(X))
```

```python
new_train_data = selector.fit_transform(X, y)
```

```python
pd.DataFrame(new_train_data).head()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
cross_val_score(LogisticRegression(), X, y, cv=10).mean()
```

```python
cross_val_score(LogisticRegression(), new_train_data, y, cv=10).mean()
```

```python
from sklearn.ensemble import RandomForestClassifier
ddos_model = RandomForestClassifier(random_state=1)
ddos_model.fit(new_train_data, y)
```

```python
ddos_model.predict(pd.DataFrame(new_train_data).head())
```
